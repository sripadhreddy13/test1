```python
df['Total'] = df[['math score', 'reading score', 'writing score']].sum(axis=1)
df['Max'] = df[['math score', 'reading score', 'writing score']].max(axis=1)
df['Min'] = df[['math score', 'reading score', 'writing score']].min(axis=1)
df['Avg'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)
```
